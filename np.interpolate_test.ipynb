```python
import numpy as np
from scipy import interpolate
import pylab as py
```

```python
x = np.r_[0:10:11j]
y = np.exp(-x/4)*x
```

```python
f = interpolate.interp1d(x,y)
```

```python
xnew = np.r_[0:10:100j]
```

```python
py.figure(1)
py.clf()
py.plot(x,y,'ro',xnew, f(xnew),'b-')
```

```python
x = np.r_[0:10:11j]
y = np.sin(x)
```

```python
for kind in ['linear','nearest', 'zero', 'slinear', 'quadratic', 'cubic']:
    f = interpolate.interp1d(x,y,kind=kind)
    ynew = f(xnew)
    py.plot(xnew,ynew,label=kind)
py.legend(loc='lower right')
```
